```python

# plt.rc('text', usetex='false') 
from sklearn.datasets import *
from sklearn import tree
from dtreeviz.trees import *
classifier = tree.DecisionTreeClassifier(max_depth=3)  # limit depth of tree
iris = load_iris()
classifier.fit(iris.data, iris.target)
feature_names=['萼片sepal length (cm)',
 'sepal width (cm)',
 '花瓣petal length (cm)',
 '花瓣petal width (cm)']
class_names=["山鸢尾 (부채붓꽃)", "变色鸢尾(ブルーフラッグ)", "维吉尼亚鸢尾(Iris-virginica)"]
viz = dtreeviz(classifier,
               iris.data,
               iris.target,
               target_name='variety',
               feature_names=feature_names,
               class_names=class_names  # need class_names for classifier
              )

viz
```
